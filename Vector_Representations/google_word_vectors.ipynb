```python
from gensim.models import KeyedVectors 
```

```python
google_vec_file = 'data/GoogleNews-vectors-negative300.bin'
```

```python
model = KeyedVectors.load_word2vec_format(google_vec_file, binary = True)
```

```python
# List the first 10 words in the vocabulary
# list(model.vocab.keys())[:10]
first_10_words = list(model.key_to_index.keys())[:50]
print(first_10_words)
```
